# Plot rocket events from the short pickle files I am sharing with Paul and Aiden

```python
import os
import header
paths = header.setup_environment()
import obspy
print(paths)
EVENTS_DIR = os.path.join(paths['outdir'], 'EVENTS')
import glob
pklfiles = sorted(glob.glob(os.path.join(EVENTS_DIR, '*_short.pkl')))
for pklfile in pklfiles:
    st = obspy.read(pklfile, format='pickle')
    stZ = st.select(component='Z')
    stZ.plot(equal_scale=False);
```
