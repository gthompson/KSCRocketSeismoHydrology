# Processing 1R miniseed files

We have previously created day-long miniseed files for the PASSCAL 1R network
Whether they are correct, we do not know.
But let us loop through all these archives, and a least create an SDS directory from them

## 1. Copy files from REFTEK_DATA/DAYS0/

```python
# modified this because I forgot the ".D" in the copied-over filename first time around
# so trying to eliminate these misnamed files

import os
import glob
network = '1R'
INPUTDIR = '/data/KSC/duringPASSCAL/REFTEK_DATA/DAYS0/' + network
for STATIONDIR in sorted(glob.glob(os.path.join(INPUTDIR, 'BHP*'))):
    print(STATIONDIR)
    for f in sorted(glob.glob(os.path.join(STATIONDIR, 'BHP*'))):
        station, network2, location, channel, yyyy, julday = os.path.basename(f).split('.')
        sdsfolder = f'/data/SDS/{yyyy}/{network}/{station}/{channel}.D'
        if not os.path.isdir(sdsfolder):
            os.makedirs(sdsfolder)

        sdsfile = '.'.join([network, station, location, channel + '.D', yyyy, julday])
        misnamedfile = '.'.join([network, station, location, channel, yyyy, julday])
        sdsfullpath = os.path.join(sdsfolder, sdsfile)
        misnamedpath = os.path.join(sdsfolder, misnamedfile)
        print(sdsfolder, sdsfile)
        
        copyf = True
        
        if os.path.isfile(sdsfullpath):
            # check sizes
            if os.path.getsize(f) > os.path.getsize(sdsfullpath):
                if os.path.isfile(misnamedpath):
                    if os.path.getsize(misnamedpath) == os.path.getsize(f):
                        cmd = f"mv {misnamedpath} {sdsfullpath}"
                        print(cmd)
                        os.system(cmd)
                        copyf = False

        if copyf:
            cmd = f"cp {f} {sdsfullpath}"
            print(cmd)
            os.system(cmd)
            
        if os.path.isfile(misnamedpath):
            if os.path.getsize(misnamedpath) > os.path.getsize(sdsfullpath):
                cmd = f"mv {misnamedpath} {sdsfullpath}"
                print(cmd)
                os.system(cmd)
            else:
                cmd = f"rm {misnamedpath}"
                print(cmd)
                os.system(cmd)
```

## 2. Copy files from REFTEK_DATA/obsolete/old/DAYFILES


```python
import os
import glob
alreadyexists=[]
network = '1R'
INPUTDIR = '/data/KSC/duringPASSCAL/REFTEK_DATA/obsolete/old/DAYFILES'
for YEARDIR in sorted(glob.glob(os.path.join(INPUTDIR, '20*'))):
    for JDAYDIR in sorted(glob.glob(os.path.join(YEARDIR, '[0-9][0-9][0-9]'))):
        for f in sorted(glob.glob(os.path.join(JDAYDIR, '1R.*'))):
            print(f)
            network, station, location, channel, typecode, yyyy, julday = os.path.basename(f).split('.')
            sdsfolder = f'/data/SDS/{yyyy}/{network}/{station}/{channel}.D'
            if not os.path.isdir(sdsfolder):
                os.makedirs(sdsfolder)
                pass
            sdsfile = os.path.basename(f)
            sdsfullpath = os.path.join(sdsfolder, sdsfile)
            print(sdsfolder, sdsfile)
            if not os.path.isfile(sdsfullpath):
                cmd = f"cp {f} {sdsfullpath}"
                print(cmd)
                os.system(cmd)
                #pass
            else:
                print(f'{sdsfullpath} already exists')
                alreadyexists.append(sdsfullpath)
print(alreadyexists)        
```
