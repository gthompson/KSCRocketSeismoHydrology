# Simple R program to loop over *.csv.bz2 files and read them

## 1. Generate a list of *.csv.bz2 files

```r
# Define the root directory where the search should begin
root_directory <- "/data/KSC/EROSION/DropboxFolder/2022_DATA/WellData/MASTERED"

# Use list.files() to get all CSV files recursively in the directory
csv_files <- list.files(root_directory, pattern = "\\.bz2$", full.names = TRUE, recursive = TRUE)

print(csv_files)
flush.console()
```

## 2. Read the first 10 rows of each file, and display them

```r
# Loop through each CSV file and read them into a list
data_list <- lapply(csv_files, function(file) {
  cat("Reading file:", file, "\n")
  data <- read.csv(file, nrows=10)  # You can customize this part (e.g., use read.csv2(), or specify options)
  print(head(data))
  flush.console()
})
```
