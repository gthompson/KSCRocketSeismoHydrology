```python
# inspect daily
import os
import pandas as pd
import header
paths = header.setup_environment()
paths['daily']=os.path.join(paths['WELLDATA_TOP'], '07_DAILY_WLM')
date=pd.to_datetime('2022-07-23')
df = pd.read_pickle(os.path.join(paths['daily'], f'{date.strftime("%Y%m%d")}.pkl'))
display(df.columns)
```

```python
import matplotlib.pyplot as plt

digits_columns = [col for col in df.columns if col.endswith('digits')]
print(digits_columns)
df.iloc[::100].plot(x='datetime', y=digits_columns, ylabel='digits')
plt.show()

calibrated_columns = [col for col in df.columns if col.endswith('calibrated')]
print(calibrated_columns)
df.iloc[::100].plot(x='datetime', y=calibrated_columns, ylabel='PSI calibrated')
plt.show()

corrected_columns = [col for col in df.columns if col.endswith('corrected')]
print(corrected_columns)
df.iloc[::100].plot(x='datetime', y=corrected_columns, ylabel='PSI corrected')
plt.show()

wlm_columns = [col for col in df.columns if col.endswith('wlm')]
print(wlm_columns)
df.iloc[::100].plot(x='datetime', y=wlm_columns, ylabel='water level (m)')
plt.show()
```
