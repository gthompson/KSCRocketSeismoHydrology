```python
# inspect daily
import os
import pandas as pd
import header
paths = header.setup_environment()
paths['daily']=os.path.join(paths['WELLDATA_TOP'], '05_DAILY_PSI')
date=pd.to_datetime('2022-11-01')
df = pd.read_pickle(os.path.join(paths['daily'], f'{date.strftime("%Y%m%d")}.pkl'))
display(df.columns)
```

```python
digits_columns = [col for col in df.columns if col.endswith('digits')]
print(digits_columns)
df.iloc[::10].plot(x='datetime', y=digits_columns, ylabel='digits')

calibrated_columns = [col for col in df.columns if col.endswith('calibrated')]
print(calibrated_columns)
df.iloc[::10].plot(x='datetime', y=calibrated_columns, ylabel='PSI calibrated')

corrected_columns = [col for col in df.columns if col.endswith('corrected')]
print(corrected_columns)
df.iloc[::10].plot(x='datetime', y=corrected_columns, ylabel='PSI corrected')
```
