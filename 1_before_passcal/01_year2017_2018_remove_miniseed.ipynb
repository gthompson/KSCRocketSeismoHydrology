## 2. We have continuous data from 2017 that is an SDS archive, but sometimes has .miniseed on end

This is for AM and FL networks

I need to copy the data from:

Input pattern: /data/KSC/beforePASSCAL/CONTINUOUS/2017/NETWORK/STATION/CHANNEL.D/NETWORK.STATION.LOCATION.CHANNEL.D.YYYY.DD[.miniseed]

to

Output pattern: /data/SDS/2017/NETWORK/STATION/CHANNEL.D/NETWORK.STATION.LOCATION.CHANNEL.D.YYYY.DD

See also: https://www.seiscomp.de/seiscomp3/doc/applications/slarchive/SDS.html


```python
import os
import glob

alreadyexists = []

TOP_INPUT_DIR = '/data/KSC/beforePASSCAL/CONTINUOUS'
for YYYYDIR in sorted(glob.glob(os.path.join(TOP_INPUT_DIR, '201[78]'))):
    yyyy = os.path.basename(YYYYDIR)
    for NETWORKDIR in sorted(glob.glob(os.path.join(YYYYDIR, '[A-Z][A-Z]'))):
        network = os.path.basename(NETWORKDIR)
        for STATIONDIR in sorted(glob.glob(os.path.join(NETWORKDIR, '[A-Z]*'))):
            station = os.path.basename(STATIONDIR)
            for CHANNELDIR in sorted(glob.glob(os.path.join(STATIONDIR, '*.D'))):
                channel = os.path.basename(CHANNELDIR).replace('.D','')
                print(CHANNELDIR, yyyy, network, station, channel)
                for f in sorted(glob.glob(os.path.join(CHANNELDIR, f"{network}.{station}.*"))):
                    sdsfolder = f'/data/SDS/{yyyy}/{network}/{station}/{channel}.D'
                    if not os.path.isdir(sdsfolder):
                        os.makedirs(sdsfolder)
                        #pass
                    sdsfile = os.path.basename(f).replace('.miniseed', '')
                    sdsfullpath = os.path.join(sdsfolder, sdsfile)
                    print(sdsfolder, sdsfile)
                    if not os.path.isfile(sdsfullpath):
                        cmd = f"cp {f} {sdsfullpath}"
                        print(cmd)
                        os.system(cmd)
                        #pass
                    else:
                        print(f'{sdsfullpath} already exists')
                        alreadyexists.append(sdsfullpath)
print(alreadyexists)
```
